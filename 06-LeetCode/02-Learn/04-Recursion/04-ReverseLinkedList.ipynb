```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None
    

class LinkedList:
    
    def __init__(self):
        self.headval = None
        
    def printList(self):
        currentPrintVal = self.headval
        while currentPrintVal is not None:
            print(" --> ", currentPrintVal.val,end="")
            currentPrintVal = currentPrintVal.next
            
    def addAtTail(self, dataVal):
        newNode = Node(dataVal)
        if self.headval == None:
            self.headval = dataVal
            return 
        else:
            currentDataVal = self.headval
            while currentDataVal.next:
                currentDataVal = currentDataVal.next
            currentDataVal.next = newNode
            
    def addAtHead(self, dataVal):
        newNode = Node(dataVal)
        newNode.next = self.headval
        self.headval = newNode
        
    def addAtIndex(self, index, dataVal):
        count = 0
        newData = Node(dataVal)
        if index <= 0:
            self.addAtHead(dataVal)
        else:
            currentNode = self.headval
            while count < index-1:
                currentNode = currentNode.next
                count += 1
            temp = currentNode.next
            currentNode.next = newData
            newData.next = temp
            
    def delHead(self):
        current = self.headval
        self.headval = current.next
        
    def deleteNode(self, removeNode):
        headval = self.headval
        
        if headval is not None:
            if headval.val == removeNode:
                self.headval = headval.next
                headval = None
                return
            
        while headval is not None:
            if headval.val == removeNode:
                break
            prev = headval
            headval = headval.next
            
        if headval is None:
            return
        
        prev.next = headval.next
        headval = None
llist = LinkedList()
llist.addAtHead("3")
llist.addAtHead("2")
llist.addAtHead("1")
llist.addAtTail("5")
llist.addAtIndex(2,"6")
llist.delHead()
llist.deleteNode("2")
print("\nInitial Linked List\n")
llist.printList()
```
