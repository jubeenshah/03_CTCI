```python
stringToPass = "12331231231312"
import re
def stringToIntegerWithRegex(string):
    m = re.search(r"^(([+-]\d+)|\d+)", string.strip())
    if m:
        ans = int(m.group(1))
    else:
        ans = 0
    if ans < -2**31:
        return -2**31
    elif ans > 2**31 - 1:
        return 2**31 - 1
    else:
        return ans
    
stringToIntegerWithRegex(stringToPass)
            
```
