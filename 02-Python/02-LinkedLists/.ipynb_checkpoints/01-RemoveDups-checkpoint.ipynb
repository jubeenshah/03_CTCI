# Remove Duplicates

Write code to remove duplicates from an unsorted linkedlist

```python
duplicateCounter = 0
duplicateDictionary = dict()
class Node:
    def __init__(self, dataVal):
        self.dataVal = dataVal
        self.nextVal = None

class singlyLinkedList:
    
    def __init__(self):
        self.headVal = None
    
    def printList(self):
        printVal = self.headVal
        while printVal is not None:
            print(printVal.dataVal)
            printVal = printVal.nextVal
            
    def insertAtBeg(self, newData):
        newNode = Node(newData)
        newNode.nextVal = self.headVal
        self.headVal = newNode
        
    def deleteNode(self, removeNode):
        headVal = self.headVal
        
        if headVal is not None:
            if headVal.dataVal == removeNode:
                self.headVal = headVal.nextVal
                headVal = None
                return
            
        while headVal is not None:
            if headVal.dataVal == removeNode:
                break
            prev = headVal
            headVal = headVal.nextVal
            
        if headVal is None:
            return
        
        prev.nextVal = headVal.nextVal
        headVal = None
        
    def countDuplicates(self):
        currentNode = self.headVal
        while currentNode.nextVal is not None:
            if currentNode.dataVal in duplicateDictionary:
                duplicateCounter += 1
            else:
                duplicateCounter = 1
            duplicateDictionary[currentNode.dataVal] = duplicateCounter
            currentNode = currentNode.nextVal
        if currentNode.dataVal in duplicateDictionary:
            duplicateDictionary[currentNode.dataVal] += 1
        else:
            duplicateDictionary[currentNode.dataVal] = 1
        return duplicateDictionary
    
    def removeDuplicates(self):
        string = ""
        for item,values in duplicateDictionary.items():
                if values > 1:
                    self.deleteNode(item)
        return string
    
list1 = singlyLinkedList()
n1 = Node("1")
n2 = Node("2")
n3 = Node("2")
n4 = Node("7")
n5 = Node("4")
n6 = Node("6")
n7 = Node("4")
list1.headVal = n1
n1.nextVal = n2
n2.nextVal = n3
n3.nextVal = n4
n4.nextVal = n5
n5.nextVal = n6
n6.nextVal = n7


print("\nCurrent Linked List\n")
list1.printList()

list1.deleteNode("5")

print("\nupdated Linked List\n")
list1.printList()

print("\nDuplicates Dictionary\n")
x = list1.countDuplicates()
list1.removeDuplicates()
print(x)
#list1.printList()

print("\nRemoving duplicates\n")
list1.printList()
```
